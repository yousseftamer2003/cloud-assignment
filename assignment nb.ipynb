```python
import pandas as pd
import datetime as dt
import numpy as np
import warnings
```

```python
d = pd.read_csv('books.csv')
d
```

```python
d.tail()
d.info()
```

```python
d['original_publication_year'] = pd.to_datetime(d['original_publication_year']) 
print(d['original_publication_year'])
```

```python
d = d[(d.books_count>=0)&(d.ratings_count>=0)&(d.work_ratings_count>=0)&(d.work_text_reviews_count>=0)&(d.ratings_1>=0)&(d.ratings_2>=0)&(d.ratings_3>=0)&(d.ratings_4>=0)&(d.ratings_5>=0)]
print(d)
```

```python
d = d.drop_duplicates()
print(d)
```

```python
d.isnull().sum()
```

```python
d.dropna(inplace=True)
d.isnull().sum()
```

```python
harry_potter_d = d[d['original_title'].str.contains('Harry Potter', case=False)]
print (harry_potter_d)
```

```python
MostSellingBooks = harry_potter_d.loc[harry_potter_d['best_book_id'].idxmax()]
print (MostSellingBooks)
```

```python
AverageRating = harry_potter_d['average_rating'].mean()
print (AverageRating)
```
